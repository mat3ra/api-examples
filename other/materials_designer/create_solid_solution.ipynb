# Create Solid Solution (Mixed Stoichiometry)

Create a solid solution by partially substituting one element with another
(e.g., Hf₁₋ₓZrₓO₂). Automatically determines optimal supercell size to achieve
the target concentration.

<h2 style="color:green">Usage</h2>

1. Make sure to select Input Materials (in the outer runtime) before running the notebook.
1. Set parameters in cell 1.1. below.
1. Click "Run" > "Run All" to run all cells.
1. Scroll down to view results.

## Notes

1. For more information, see [Introduction](Introduction.ipynb)

## 1. Prepare the Environment
### 1.1. Set solid solution parameters

```python
SOURCE_ELEMENT = "Ni"       # Element to partially replace
TARGET_ELEMENT = "Co"       # Replacement element
CONCENTRATION = 0.5        # Fraction of SOURCE_ELEMENT sites to replace (0–1)

# Site selection: "random" or "uniform" (Farthest Point Sampling for even spacing)
SITE_SELECTION_METHOD = "uniform"

# Random seed for reproducible site selection
SEED = 0

# Tolerance for matching target concentration (smaller = larger supercell)
TOLERANCE = 0.01
```

### 1.2. Install Packages
The step executes only in Pyodide environment. For other environments, the packages should be installed via `pip install` (see [README](../../README.ipynb)).

```python
from mat3ra.notebooks_utils.packages import install_packages

await install_packages("made")
```

### 1.3. Get input materials

```python
from mat3ra.notebooks_utils.material import get_materials

materials = get_materials(globals())
```

## 2. Create Solid Solution

```python
from mat3ra.made.tools.helpers import create_solid_solution

unit_cell = materials[0]
solid_solution = create_solid_solution(
    material=unit_cell,
    source_element=SOURCE_ELEMENT,
    target_element=TARGET_ELEMENT,
    concentration=CONCENTRATION,
    seed=SEED,
    tolerance=TOLERANCE,
    site_selection_method=SITE_SELECTION_METHOD,
)

from collections import Counter
composition = Counter(solid_solution.basis.elements.values)
n_cation = composition.get(SOURCE_ELEMENT, 0) + composition.get(TARGET_ELEMENT, 0)
actual_fraction = composition.get(TARGET_ELEMENT, 0) / n_cation if n_cation else 0

print(f"Unit cell: {unit_cell.name}")
print(f"Composition: {dict(composition)}")
print(f"Total atoms: {sum(composition.values())}")
print(f"Requested concentration: {CONCENTRATION}")
print(f"Actual concentration:    {actual_fraction:.4f}")
print(f"Site selection method:   {SITE_SELECTION_METHOD}")
```

## 3. Visualize Result(s)

```python
from mat3ra.notebooks_utils.ipython.entity.material.visualize import visualize_materials as visualize

visualize(
           {"material": solid_solution, "title": f"Solid Solution: {dict(composition)}"}, viewer="wave")
```
